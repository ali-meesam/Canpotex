```python
def interim_pricing_df(std,prm):
    structure = [
        ['CanpotexPool', 'Grade', 'ProductAnalysis', 'US/MT-K2O', 'US/MT-KCL'],
        ['standard', 'Standard', 0.605, 100, 60.5],
        ['standard', 'Suspension', 0.605, 100, 60.5],
        ['premium', 'Coarse', 0.608, 200, 121.6],
        ['premium', 'Granular', 0.608, 200, 121.6],
        ['premium', 'Ind. Sol 2', 0.625, 200, 125.0],
        ['premium', 'Soluble', 0.625, 200, 125.0],
        ['premium', 'Ref/Soft', 0.63, 200, 126.0],
        ['standard', 'Wh. Standard', 0.625, 100, 62.5],
        ['premium', 'Wh. Granular', 0.625, 200, 125.0],
        ['premium', 'Chicklets', 0.625, 200, 125.0],
        ['premium', 'Prills', 0.61, 200, 122.0]
    ]
    
    d = pd.DataFrame(structure[1:],columns=structure[0])

    d['US/MT-K2O'] = d.apply(lambda x: std if x.CanpotexPool=='standard' else prm, axis=1)
    d['US/MT-KCL'] = d['US/MT-K2O']*d.ProductAnalysis
    return d
```

```python
interim_pricing_df(*interim_k20.iloc[-1][['Standard K20 Price','Premium K20 Price']].values.tolist())
```
